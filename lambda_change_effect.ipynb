# Matplotlib Figures

```python
import pandas as pd
import matplotlib.pyplot as plt 


df = pd.read_csv("feature_lambda_cora.csv")
lambda_val = [0,0.25,0.5,0.75,1.0,1.25,1.5,1.75,2.0]
combined = df['combined'].tolist()
contextual = df['contextual'].tolist()
structural = df['structural'].tolist()
joint_type = df['joint_type'].tolist()
structure_type = df['structure_type'].tolist()


plt.plot(lambda_val, contextual,   label = 'Contextual', marker='x', linewidth=1)
plt.plot(lambda_val, structural,   label = 'Structural', marker='o', linewidth=1)
plt.plot(lambda_val, joint_type,   label = 'Joint-type', marker='o', linewidth=1)
plt.plot(lambda_val, combined,   label = 'Contextual + Structural', marker='o', linewidth=1)
plt.plot(lambda_val, structure_type,   label = 'Structural + Joint-type', marker='o', linewidth=1)


# plt.xlabel('Feature Reconstruction λ')
plt.ylabel('ROC_AUC')
plt.xlabel(r"Varying $\lambda^'_x$")
plt.legend(loc='lower right')
plt.xticks(lambda_val)
plt.yticks([20,40,60,80,100])
plt.title('Impact of feature reconstruction loss weight on Cora')
plt.savefig('./feature_lambda_cora.pdf',bbox_inches='tight')
plt.show()
```
